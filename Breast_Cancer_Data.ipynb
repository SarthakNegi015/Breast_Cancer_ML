# Splitting the Dependent and Independent Variables

```python
data1
```

```python
x=data1.iloc[:,:-1]
```

```python
x
```

```python
y=data1.iloc[:,-1]
```

```python
y.head()
```

# Train Test Splitting the DataSet

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

# Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred_y=model.predict(x_test)
```

```python
pred_y
```

```python
np.array(y_test)
```

```python
sns.(pred_y-y_test)
```

# Checking the accuracy

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,pred_y)
accuracy
```

```python
from sklearn.metrics import recall_score
```

```python
r=recall_score(y_test,pred_y)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import fbeta_score
```

```python
a=accuracy_score(y_test,pred_y)
p=precision_score(y_test,pred_y)
f=fbeta_score(y_test,pred_y,beta=2.0)
```

```python
print(f"Result:\n precision={p} \n recall={r} \n f_beta={f} \n accuracy={a}")
```
